```python
import pandas as pd

# Load GDP Per Capita Data
gdp_df = pd.read_csv(r'YOUR_DIR')
gdp_df = gdp_df.loc[:, ~gdp_df.columns.str.contains('^Unnamed')]  # Drop unnamed columns
gdp_df.fillna(0, inplace=True)
gdp_df_long = pd.melt(gdp_df, id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], 
                      var_name='Year', value_name='GDP per capita (current US$)')

# Load Tourism Data
tourism_df = pd.read_csv(r'YOUR_DIR')
tourism_df = tourism_df.loc[:, ~tourism_df.columns.str.contains('^Unnamed')]  # Drop unnamed columns
tourism_df.fillna(0, inplace=True)
tourism_df_long = pd.melt(tourism_df, id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], 
                          var_name='Year', value_name='International tourism, number of arrivals')  # âœ… Use full name

# Merge only necessary columns
merged_df = pd.merge(gdp_df_long[['Country Name', 'Country Code', 'Year', 'GDP per capita (current US$)']],
                     tourism_df_long[['Country Name', 'Country Code', 'Year', 'International tourism, number of arrivals']],
                     on=['Country Name', 'Country Code', 'Year'],
                     how='outer')

# Clean Year column
merged_df = merged_df[pd.to_numeric(merged_df['Year'], errors='coerce').notna()]
merged_df['Year'] = merged_df['Year'].astype(int)

# Save the merged data
output_path = r'YOUR_DIR'
merged_df.to_csv(output_path, index=False)

# Display basic summary
print(merged_df.head())
print(merged_df.describe())
```
